### Convert emp_name to proper case ---upper /lower ---Initcap (CamelCase)

```sql
Select emp_id , upper(emp_name) as upper_emp_name , initcap(emp_name) as Corrected_name , department , base_salary , bonus , joining_date
From employee_payments
```

### Calculate total income = base_salary + bonus (NULL safe) 

```sql
SELECT
       emp_name,
       base_salary,
       bonus,

       COALESCE(base_salary,0) + COALESCE(bonus,0) AS total_income

FROM employee_payments;
```

```sql
select emp_name , Round(coalesce(base_salary,0) + coalesce(bonus,0),0) as Rounded_income from employee_payments
```

### Extract joining year

```sql
Select emp_name , Extract(year from joining_date) as joining_year from employee_payments
```

### Use CASE to classify: Senior if experience > 7 years , Mid if between 4 and 7 ,Junior otherwise


```sql
Select emp_name , joining_date , timestampdiff(year , joining_date , current_date) as experience , 
case 
when timestampdiff(year , joining_date , current_date) > 7 then 'Senior'
when timestampdiff(year , joining_date , current_date) between 4 and 7 then 'Intermediate'
else 'Junior'
end as employee_level
from employee_payments
```

### Use CASE: Same-day , Delayed (>3 days) , Pending

```sql
SELECT
       order_id,
       order_date,
       delivery_date,

       CASE
            WHEN delivery_date IS NULL THEN 'Pending'

            WHEN DATEDIFF(delivery_date, order_date) = 0
            THEN 'Same-day'

            WHEN DATEDIFF(delivery_date, order_date) > 3
            THEN 'Delayed (>3 days)'

            ELSE 'On-time'
       END AS delivery_status

FROM orders;
```

### Monthly interest using power function

```sql
SELECT
    loan_id,
    customer_name,
    loan_amount,
    interest_rate,

    round(loan_amount * interest_rate * POWER(12, -1)/ 100 , 2) 
        AS monthly_simple_interest

FROM loan_details;
```

### Years since loan start

```sql
SELECT 
    loan_id,
    customer_name,
    loan_start,

    ROUND(
        months_between(current_date(), loan_start) / 12,
        2
    ) AS years_since_loan_start

FROM loan_details;
```

### · CASE: Excellent ≥90% , Average 75–89% , Poor otherwise


```sql
SELECT 
    emp_id,
    emp_name,

    ROUND(present_days / total_days * 100, 2) AS attendance_percentage,

    CASE 
        WHEN ROUND(present_days / total_days * 100, 2) >= 90 THEN 'Excellent'
        WHEN ROUND(present_days / total_days * 100, 2) BETWEEN 75 AND 89 THEN 'Average'
        ELSE 'Poor'
    END AS attendance_status

FROM attendance;
```
